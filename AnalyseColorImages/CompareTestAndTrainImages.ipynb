<h1 style="text-align: center;">Compare images in train and test set to make sure all images are unique</h1>

```python
# import basics
import os

# custom modules
import sys
sys.path.append('..')
from vision.Utils import info
from vision.Definitions import *
from vision.Utils import readDataToDataFrame, createDirectories
from vision.TransformImages import TransformImages
from vision.Data import Data
# MIN_SIZE_IMAGE is 224, defined in vision.Definitions

pathDataSet = "datasets"
pathDataSet224x224 = f"datasets{MIN_SIZE_IMAGE[0]}x{MIN_SIZE_IMAGE[1]}"
```

<h1>1. Preprocess the data</h1>

### Create the DataFrames

```python
# "training_set" and "test_set" are from https://www.kaggle.com/datasets/charmz/cats-and-dogs
df_train = readDataToDataFrame(os.path.join(pathDataSet, "training_set"))
df_test = readDataToDataFrame(os.path.join(pathDataSet, "test_set"))
```

### Create the DataSets

```python
dataSetTrain = Data(df_train)
dataSetTest = Data(df_test)
```

<h1>2. Filter the data </h1>

### Create a new directory and resize the images to MIN_SIZE_IMAGE

```python
# create a new directory with subfolders for images with size MIN_SIZE_IMAGE
folders = ["training_set", 
           "test_set"]
    
# resize the images and save it in the new directory             
if createDirectories(pathDataSet224x224, folders):
    print(f"Resizing images")
    transformer = TransformImages(pathDataSet, pathDataSet224x224, resize=True, size=MIN_SIZE_IMAGE)
    transformer.resize() 
```

### Create a new dataframe with equal image size MIN_SIZE_IMAGE

```python
df_train    = readDataToDataFrame(os.path.join(pathDataSet224x224, "training_set"))
df_test     = readDataToDataFrame(os.path.join(pathDataSet224x224, "test_set"))
```

### Check the new data frames. We see that all small images have equal size of 224x224 pixel.

```python
df_train.describe()
```

```python
df_test.describe()
```

### Create a new DataSets with equal image size MIN_SIZE_IMAGE

```python
dataSetTrain = Data(df_train)
dataSetTest = Data(df_test)
```

<h1>3. Compare the images in the train and test set </h1>

```python
# calculate the mean square error between the images by using cv.norm
import cv2 as cv

similarImages: int = 0
numImagesTrain = len(df_train)
    
for i in df_test.index:
    im_test_filename = df_test['fileName'][i]
    im_test = cv.imread(im_test_filename)
    im_test_width = df_test['width'][i]
    im_test_height = df_test['height'][i]
    
    if not i % 100:
        print(f"Compared {i} testing images with {numImagesTrain} training images.")
    
    for j in df_train.index:
        im_train_filename = df_train['fileName'][j]
        im_train = cv.imread(im_train_filename)
        im_train_width = df_train['width'][j]
        im_train_height = df_train['height'][j]
        
        assert im_train_width == im_test_width
        assert im_train_height == im_test_height
        
        error_l2 = cv.norm(im_test, im_train, cv.NORM_L2)
        similarity = 1 - error_l2 / (im_test_height * im_test_width)
        
        if similarity > 0.80:
            similarImages+=1
            print(f"similarity: {similarity} for {im_test_filename} and {im_train_filename}")
            # use the dataset for plotting, the index in the dataset and dataframe is equal
            dataSetTest.plotImages([i])
            dataSetTrain.plotImages([j])
            
print(f"Found {similarImages} similar images!")
```
